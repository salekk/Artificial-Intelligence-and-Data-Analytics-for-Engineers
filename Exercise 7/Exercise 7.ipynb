```python
data = pd.read_csv('store_data.csv', header=None)
data
```

```python
store = []
for i in range (0,7501):
    store.append([str(data.values[i,j]) for j in range (0,20)])
store
```

d) Apply the Apriori algorithm and store the result into a variable, e.g. association_rules. It

```python
association_rules = apriori(transactions = store, min_support=0.0045, min_confidence=0.2, min_lift=3,
min_length=2)
```

## Task 2 - Dimensionality Reduction

In this task, you will use unsupervised learning to implement a dimensionality reduction from one of our
previous lectures (“PCA”). 

For this task we’ll use the MNIST database of handwritten digits. It has a training set of 60,000 examples, and a test set of 10,000 examples. It is a subset of a larger set available from NIST.

The digits have been size-normalized and centered in a fixed-size image. Each image is 28x28 pixels, thus, each observation has 784 dimensions.

a) Import the dataset “MNIST original” from the datasets potion of sklearn
(sklearn.datasets.fetch_mldata)

```python
from sklearn.datasets import fetch_openml
X, y = fetch_openml('mnist_784', version = 1, return_X_y=True)
```

```python
import matplotlib.pyplot as plt
plt.imshow(X[0].reshape((28,28)))
plt.show()
```

b) Split the data using the train_test_split method

```python
from sklearn.model_selection import train_test_split

X,y = X/255., y.astype(float)
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size=0.20, random_state=42)
```

c) Import the following libraries (which will be used for visualization):


```python
from time import time
import numpy as np
from sklearn import manifold
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib import offsetbox
```

d) Import the following libraries (which will be used for visualization):

```python
def plot_embedding(X,y, title='None'):
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)
    
    plt.figure()
    ax = plt.subplot(111)
    for i in range(X.shape[0]):
        plt.text(X[i, 0], X[i, 1], str(y[i]),
            color=plt.cm.Set1(y[i] / 10.),
            fontdict={'weight': 'bold', 'size': 9})
        
        
    plt.xticks([]), plt.yticks([])
    if title is not None:
        plt.title(title)
        
    
indexes=np.random.choice(len(X_train),size=300)
```

e) Train an embedding model (PCA, umap) and plot the result. Try to explain the purpose of each code
line and the resulting visualization.

```python
embedding_PCA = PCA(n_components=2)
t0 = time()
X_embedding = embedding_PCA.fit_transform(X_train)

plot_embedding(X_embedding[indexes],y_train[indexes],
              "Embedding of the digits (time %.2f)"%
              (time()-t0))
plt.show()
```
